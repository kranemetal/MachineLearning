### K-NN Notes
#### Steps
1. Choose the number K of neighbors(5 is commonly accepted), to be counted for a given datapoint.
2. Take the K nearest neighbors of the new data point, according to the Euclidean(or any other) distance.
3. Among these K neighbors, count the number of data points in eacg category.
4. Assign the new data point to the category where you counted the most neighbors.
5. Model is ready.

### Splitting independent variable X and dependent variable Y

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

### Splitting the train and test datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

### Train K-NN model on train set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p=2) 
classifier.fit(x_train, y_train) 
# n_neighbors is k, minkowski & p=2 means Euclidean distance, minkowski & p=1 means Manhattan distance,
# Manhattan distance is city-block distance or taxi-cab distance
# Euclidean distance is geometric distance(straight line) in Euclidean space
```

### Predicting a new result
For a customer aged 35 years and earning USD 82k

```python
print(classifier.predict(sc.transform([[35,82000]])))
```

### Predicting the test results

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Checking the confusion matrix and accuracy score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred) #confusion matrix(actual, predicted)
print(cm)
accuracy_score(y_test, y_pred) #accuracy score (actual, predicted)
```

<b>Note:</b> Here correct predictions from two classes combined is 73+37=110 and incorrect predictions from two classes combined is 6+4=10.<br>
    Hence, the accuracy is 110/(110+10=120) = 91%.
