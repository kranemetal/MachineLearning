### Support Vector Machine(SVM) Classification Notes
- If we have data points say from 2 different classes, than we need a line separating them i.e. line of decision boundary.
- This line would help us to classify the future data points into one of these 2 classes.
- The line that separates the classes with maximum margin from both sides, is the one we need in our model.
- It should have maximum distance from vectors of both sides.

### Why is SVM different then other classifiers

- Generally any ML algorithm learns from the data-points which have typical characteristics of their class, so as to know them properly and make more accurate predictions for new/future data points.
- For example, any ML algorithm will learn from the typical points of Class A and Class B respectively, to accurately predict future data points into one of these classes.
- But SVM is different, it actually learns from the Class A data points which are similar to Class B and will also learn from Class B data points which have characteristics similar to Class A.
- Such data points (vectors) are then used as support vectors and further classifications are done based on them.

### Splitting independent variable X and dependent variable Y

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

### Splitting the train and test datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

### Train SVM model on train set

```python
from sklearn.svm import SVC
classifier = SVC (kernel = 'linear', random_state=0) 
classifier.fit(x_train, y_train) 
```

### Predicting a new result
For a customer aged 38 years and earning USD 86k

```python
print(classifier.predict(sc.transform([[38,86000]])))
```

### Predicting the test results

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Checking the confusion matrix and accuracy score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred) #confusion matrix(actual, predicted)
print(cm)
accuracy_score(y_test, y_pred) #accuracy score (actual, predicted)
```

<b>Note:</b> Here correct predictions from two classes combined is 75+29=104 and incorrect predictions from two classes combined is 4+12=16.<br>
    Hence, the accuracy is 104/(104+16=120) = 86%.
